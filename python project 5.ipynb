EQUIVALENT STOCK PRICE GRAPH FOR CONCATENATED STOCKS PRICE DATA OF STOCKS LISTED IN NASDAQ:
[1] WE WILL MAKE THE SECOND TEST CASE USING {UPST,PFE,ADBE}
[2] WE WILL MAKE THE FIRST TEST CASE USING {AAPL,GOOG,AMZN}


```python
import datetime as dt 
import pandas as pd 
import pandas_datareader as pdr
import matplotlib.pyplot as plt

stocks = ['UPST','PFE','ADBE']
api_key = 'YOUR_ALPHA_VANTAGE_API_KEY'  
end = dt.datetime.now()
start = end - dt.timedelta(days=5000)

# Create an empty DataFrame to store the concatenated data
concatenated_df = pd.DataFrame()

for stock in stocks:
    # Retrieve data for each stock
    stock_data =  pdr.av.time_series.AVTimeSeriesReader(symbols=stock, api_key=api_key).read()
    stock_data['Symbol'] = stock  # Add a 'Symbol' column to identify the stock
    concatenated_df = pd.concat([concatenated_df, stock_data])

# Print the head of the concatenated DataFrame
print(concatenated_df.head())

plt.figure(figsize=(10, 6))

for stock in stocks:
    # Filter data for each stock
    stock_close = concatenated_df[concatenated_df['Symbol'] == stock]['close']
    plt.plot(stock_close, label=stock)

plt.title('Stock Closing Prices')
plt.xlabel('Date')
plt.ylabel('Closing Price')
plt.legend()
plt.grid(True)
plt.tight_layout()
plt.show()
```

```python
import datetime as dt
import yfinance as yf
import matplotlib.pyplot as plt

stocks = ['AAPL', 'GOOG', 'AMZN']
start_date = dt.datetime.now() - dt.timedelta(days=5000)  # Adjust as needed

# Create an empty DataFrame to store the concatenated data
concatenated_df = pd.DataFrame()

for stock in stocks:
    # Retrieve data for each stock from Yahoo Finance
    stock_data = yf.download(stock, start=start_date)
    stock_data['Symbol'] = stock  # Add a 'Symbol' column to identify the stock
    concatenated_df = pd.concat([concatenated_df, stock_data])

# Print the head of the concatenated DataFrame
print(concatenated_df.head())

plt.figure(figsize=(10, 6))

for stock in stocks:
    # Filter data for each stock
    stock_close = concatenated_df[concatenated_df['Symbol'] == stock]['Close']
    plt.plot(stock_close, label=stock)

plt.title('Stock Closing Prices')
plt.xlabel('Date')
plt.ylabel('Closing Price')
plt.legend()
plt.grid(True)
plt.tight_layout()
plt.show()
```
